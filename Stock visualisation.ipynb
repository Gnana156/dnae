```python
%matplotlib notebook

import pandas as pd
import pandas_datareader as pdr

import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
```

```python
tickers = ['AAPL', 'MSFT']
stocks = {}
for ticker in tickers:
    stocks[ticker] =  pdr.get_data_yahoo(ticker)
```

```python
f, axs = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

pd.concat(dict([(k, v['Adj Close']) for k, v in stocks.items()]),
          axis=1).plot(ax=axs[0])

pd.concat(dict([(k, v['Volume']) for k, v in stocks.items()]),
          axis=1).plot(ax=axs[1])

axs[0].set_ylabel('Share Price')
axs[1].set_ylabel('Volume')

f.suptitle('Share Price for: {}'.format(tickers), size=15)
```
